```python
import cv2

# 加载你训练好的YOLOv8模型
from ultralytics import YOLO
model = YOLO(r"D:\YOLO-badminton-code\badminton_detection\men_women_doubles\weights\best.pt")

# 打开视频文件
cap = cv2.VideoCapture(r"D:\YOLO-badminton-code\[4K50FPS] Zheng Siwei _Huang Yaqiong vs Seo Seung Jae _Chae Yu Jung _ 2024 Asia.mp4")

# 获取视频帧的宽度、高度和帧率
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# 定义输出视频文件
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(r"D:\YOLO-badminton-code\badminton_output2.mp4", fourcc, fps, (width, height))

# 循环读取每一帧
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # 使用YOLOv8模型对每一帧进行预测
    results = model(frame)

    # 获取预测结果并绘制边界框
    for result in results:
        boxes = result.boxes.xyxy.cpu().numpy()  # 获取每个边界框的坐标
        scores = result.boxes.conf.cpu().numpy()  # 获取每个检测框的置信度
        classes = result.boxes.cls.cpu().numpy()  # 获取检测到的类别
        
        for box, score, cls in zip(boxes, scores, classes):
            x1, y1, x2, y2 = map(int, box)
            label = f'{model.names[int(cls)]} {score:.2f}'
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # 写入输出视频文件
    out.write(frame)

    # 显示视频
    cv2.imshow('Badminton Detection', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 释放资源
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# 加载你训练好的YOLOv8模型
from ultralytics import YOLO
model = YOLO(r"D:\YOLO-badminton-code\badminton_detection\men_women_doubles\weights\best.pt")

# 打开视频文件
cap = cv2.VideoCapture(r"D:\YOLO-badminton-code\[4K50FPS] Zheng Siwei _Huang Yaqiong vs Seo Seung Jae _Chae Yu Jung _ 2024 Asia.mp4")

# 保存运动员和羽毛球位置的列表
player_positions={0:[],1:[],2:[],3:[],4:[]}

# 定义场地的四个角点（需要根据具体场地进行标注确定）
pts_src = np.array([
    [640, 572],  # 左上角
    [1292, 572],  # 右上角
    [330, 1027],  # 左下角
    [1598, 1027]   # 右下角
], dtype='float32')

# 俯视图目标位置
pts_dst = np.array([
    [0, 0],
    [500, 0],
    [0, 1000],
    [500, 1000]
], dtype='float32')

# 获取透视变换矩阵
M = cv2.getPerspectiveTransform(pts_src, pts_dst)

# 获取视频帧的宽度、高度和帧率
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# 定义输出视频文件
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(r"D:\YOLO-badminton-code\badminton_output2.mp4", fourcc, fps, (width, height))

# 循环读取每一帧
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # 使用YOLOv8模型对每一帧进行预测
    results = model(frame)

    # 获取预测结果并绘制边界框
    for result in results:
        boxes = result.boxes.xyxy.cpu().numpy()  # 获取每个边界框的坐标
        scores = result.boxes.conf.cpu().numpy()  # 获取每个检测框的置信度
        classes = result.boxes.cls.cpu().numpy()  # 获取检测到的类别
        
        for box, score, cls in zip(boxes, scores, classes):
            x1, y1, x2, y2 = map(int, box)
            center_x = (x1+x2)//2
            center_y = y2
            id = int(cls)
            if id in player_positions:
                player_positions[id].append((center_x, center_y))
            label = f'{model.names[int(cls)]} {score:.2f}'
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # 创建俯视图画布
    top_view = np.zeros((1000, 500,3), dtype='uint8')
    # 将运动员的位置进行透视变换并绘制在俯视图上
    for id, positions in player_positions.items():
        if positions:
            pos = positions[-1]
            pts = np.array([[pos]], dtype='float32')
            transformed_pos = cv2.perspectiveTransform(pts, M)
            tx, ty = transformed_pos[0][0]

            # 绘制运动员在俯视图中的位置
            color = (242, 60, 86) if id == 0 else (44, 68, 244)
            cv2.circle(top_view, (int(tx), int(ty)), 5, color, -1)

    # 缩小俯视图画布以适应原始帧
    small_top_view = cv2.resize(top_view, (200, 400))

    # 将俯视图画布叠加到原始帧的右下角
    h, w, _ = frame.shape
    frame[h-400:h, w-200:w] = small_top_view
    # 写入输出视频文件
    out.write(frame)

    # 显示视频
    cv2.imshow('Badminton Detection', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 释放资源
cap.release()
out.release()
cv2.destroyAllWindows()
```
